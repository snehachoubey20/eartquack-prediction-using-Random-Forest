```python
numerical_features = train_df.select_dtypes(include = ['int64']).columns
correlations= {}
for features in numerical_features:
  correlation = train_df[features].corr(train_df['damage_grade'])
  correlations[features]= correlation
sorted_correlations = sorted(correlations.items(), key= lambda x:abs(x[1]), reverse = True)
print("Correlation of Numerical Features with Damage Grade")
for features, correlation in sorted_correlations:
  print(f"{features}:{correlation}")
```

```python
from scipy.stats import chi2_contingency
categorical_features = train_df.select_dtypes(include=['object']).columns
chi2_results = {}
for features in categorical_features:
  contingency_table = pd.crosstab(train_df[features], train_df['damage_grade'])
  chi2, p, _, _ = chi2_contingency(contingency_table)
  chi2_results[features] = chi2, p

sorted_chi2_results = sorted(chi2_results.items(), key=lambda x: x[1][1])
print("Chi-square Test Results for Categorical Features:")
for features, (chi2, p) in sorted_chi2_results:
    print(f"{features}: Chi2 = {chi2}, p-value = {p}")
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder(dtype='int')
train = train_df.copy()
test = test_df.copy()

# Fit and transform ordinal encoder on train data
train_encoded = ordinal.fit_transform(train[obj])

# Transform test data using the same ordinal encoder
test_encoded = ordinal.transform(test[obj])

# Convert the encoded arrays back into DataFrames
train_encoded_df = pd.DataFrame(train_encoded, columns=[f"{col}_encoded" for col in obj])
test_encoded_df = pd.DataFrame(test_encoded, columns=[f"{col}_encoded" for col in obj])

# Drop the original categorical columns from train and test DataFrames
train = train.drop(columns=obj)
test = test.drop(columns=obj)

# Concatenate the encoded DataFrames with the remaining columns
train = pd.concat([train, train_encoded_df], axis=1)
test = pd.concat([test, test_encoded_df], axis=1)

# Now you can use train.head() to view the first few rows of the DataFrame
train.head()

```

```python
test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
testId = test['building_id']
train_data = train.drop(['building_id'], axis=1)
test_data = test.drop(['building_id'], axis=1)
y = train.damage_grade
X = train.drop('damage_grade', axis=1)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_val)
predictions = np.round(predictions)
f1_score(y_val, predictions, average='macro')
sub_pred = model.predict(test)
sub_pred = np.round(sub_pred)
df = pd.DataFrame({'building_id':testId.values,
                   'damage_grade':sub_pred})
df['damage_grade'] = df['damage_grade'].astype(int)
```

```python
df.to_csv("submission.csv", index=False)
```
